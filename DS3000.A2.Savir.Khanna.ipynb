## Question 6: 
The neonatal unit requests a human-readable converter for charting: implement a function convert_birthweight(grams) The function converts a weight given in grams into whole pounds and integer ounces, and returns a tuple: (pounds, ounces). If grams < 0, raise ValueError("grams must be a non-negative number"). 
Hint: 1 lb = 453.59237 g, 1 lb = 16 oz; convert remainder to ounces and keep oz in 0–15.

Note: Use the convert_birthweight() function to convert the birth weight to pounds and ounces in the questions 7-11 below.

```python
def convert_birthweight(grams):
    if grams < 0:
        raise ValueError("Grams must must a non-negative number.")
    pounds = int(grams // 453.59237)
    
    gs_per_oz = 453.59237 / 16
    remainder = grams - (pounds * 453.59237)
    oz = int(remainder // gs_per_oz)
    
    return (pounds, oz)
```
