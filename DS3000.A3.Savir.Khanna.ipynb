```python
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
```

```python
lst = [item for item in cleaned_tripdata_df["trip_type"] if item == 1]
lst2 = [item for item in cleaned_tripdata_df["trip_type"] if item == 2]

plt.bar("Street-hail", len(lst))
plt.bar("Dispatch", len(lst2))

plt.xlabel("Trip Type")
plt.ylabel("Count #")
plt.title("NYC Taxis Requesting Method")

plt.show()
```

```python
days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
cleaned_tripdata_df["lpep_pickup_datetime"] = pd.to_datetime(
    cleaned_tripdata_df["lpep_pickup_datetime"], errors="coerce"
)
cleaned_tripdata_df["pickup_dow"]  = cleaned_tripdata_df["lpep_pickup_datetime"].dt.day_name()

monday_lst = [date for date in cleaned_tripdata_df["pickup_dow"] if date == days_of_week[0]]
tuesday_lst = [date for date in cleaned_tripdata_df["pickup_dow"] if date == days_of_week[1]]
wednesday_lst = [date for date in cleaned_tripdata_df["pickup_dow"] if date == days_of_week[2]]
thursday_lst = [date for date in cleaned_tripdata_df["pickup_dow"] if date == days_of_week[3]]
friday_lst = [date for date in cleaned_tripdata_df["pickup_dow"] if date == days_of_week[4]]
sat_lst = [date for date in cleaned_tripdata_df["pickup_dow"] if date == days_of_week[5]]
sun_lst = [date for date in cleaned_tripdata_df["pickup_dow"] if date == days_of_week[6]]

plt.bar(days_of_week[0], len(monday_lst))
plt.bar(days_of_week[1], len(tuesday_lst))
plt.bar(days_of_week[2], len(wednesday_lst))
plt.bar(days_of_week[3], len(thursday_lst))
plt.bar(days_of_week[4], len(friday_lst))
plt.bar(days_of_week[5], len(sat_lst))
plt.bar(days_of_week[6], len(sun_lst))

plt.xlabel("Count of taxi pickups")
plt.ylabel("Days of the Week")
plt.title("Busisest days of the week for Taxis in NYC over Nov 23")

plt.show()
```

Seems like Wednesday and Thursday are clearly the busiest days for taxis. This makes sense and people often need to get to work for meetings those days, and business travel often happens on a weekday. Also explains why (understandly) the weekends are lowest.

```python
pu_locations = dict()
for locationID in cleaned_tripdata_df["PULocationID"]:
    if locationID not in pu_locations.keys():
        pu_locations[locationID] = 1
    else:
        pu_locations[locationID] += 1

sorted_vals = sorted(pu_locations.values())
sorted_keys = []
for val in sorted_vals[-5:]:
    for key, val2 in pu_locations.items():
        if val == val2:
            sorted_keys.append(np.int64(key))

sorted_keys = np.array(sorted_keys)
labels = [str(k) for k in sorted_keys]


plt.bar(labels, sorted_vals[-5:])

plt.xlabel("Area Code")
plt.ylabel("Times picked up")
plt.title("Top 5 pickup locations in NY during Nov 2023")

plt.show()
```

The top locations for pickup are significantly higher than both the mean and every other pickup area code, suggesting these few areas have significantly more traffic.

```python
do_locations = dict()
for locationID in cleaned_tripdata_df["DOLocationID"]:
    if locationID not in do_locations.keys():
        do_locations[locationID] = 1
    else:
        do_locations[locationID] += 1

sorted_vals = sorted(do_locations.values())
sorted_keys = []
for val in sorted_vals[-5:]:
    for key, val2 in do_locations.items():
        if val == val2:
            sorted_keys.append(np.int64(key))

sorted_keys = np.array(sorted_keys)

labels = [str(k) for k in sorted_keys]


plt.bar(labels, sorted_vals[-5:])

plt.xlabel("Area Code")
plt.ylabel("Times picked up")
plt.title("Top 5 dropoff locations in NY during Nov 2023")

plt.show()
```

Similar to the pickup locations, taxis codes 74 and 75 remain the highest and 3 out of the top 5 are the same. The mjor differences are 238 and 236 appear over 82 and 166.

```python
short_df = cleaned_tripdata_df[cleaned_tripdata_df["trip_distance"] < 2]
short_min_tip = np.min(short_df["tip_amount"])
short_max_tip = np.max(short_df["tip_amount"])
short_mean_tip = np.mean(short_df["tip_amount"])

long_df = cleaned_tripdata_df[cleaned_tripdata_df["trip_distance"] > 2]
long_min_tip = np.min(long_df["tip_amount"])
long_max_tip = np.max(long_df["tip_amount"])
long_mean_tip = np.mean(long_df["tip_amount"])

print(f"The minimum and maximum tips for rides under 2 miles was ${short_min_tip} and ${short_max_tip}, with a mean of ${round(short_mean_tip, 2)}.")
print(f"The minimum and maximum tips for rides over 2 miles was ${long_min_tip} and ${long_max_tip}, with a mean of ${round(long_mean_tip, 2)}.")
print("As expected, longer trips generally have much larger tips, over twice as much on average in fact. In both cases however, the max and minimum tip were realtively similar, implying that most people tend to tip in modest ranges regardless of lengeth.")
```
