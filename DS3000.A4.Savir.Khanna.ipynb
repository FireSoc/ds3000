## Question 4: 
(5 pts) Develop a function named annualAdmissions(). This function will calculate the total reported admissions each year across the entire state of New York. Subsequently, display the results using a line chart and annotate the chart to indicate the year with the highest number of admissions.
After which, Execute the function in a new cell. Explain the chart and discuss any patterns or trends that you have observed.


```python
years = [i+2007 for i in range(15)]
total_admissions_lst = []
def annualAdmissions():
    for year in years:
        total_admissions_lst.append(df[df["Year"] == year]["Admissions"].sum())
    return total_admissions_lst
```

```python
admissions_lst_ny = annualAdmissions()
peak_val = max(admissions_lst_ny)
peak_idx = admissions_lst_ny.index(peak_val)
peak_year = years[peak_idx]

plt.annotate(
    f"Peak {peak_year}: {peak_val:,}",
    xy=(peak_year, peak_val),                # point to the peak
    xytext=(peak_year + 0.5, peak_val * 1.05),  # text a bit to the right & above
    arrowprops=dict(arrowstyle="->", lw=1),
    ha="left", va="bottom"
)
plt.plot(years, admissions_lst_ny)

plt.xlabel("Years")
plt.ylim(0, 400000)
plt.ylabel("Admissions")
plt.title("Admissions per Year in New York State")

plt.show()
```

Admissions soft peak around 2009, and then gentle decline/stability until 2019. 2020 however, the admissions dropped precipitously, which can likely be attributed to the COVID-19 outbreak.

## Question 5: 
(10 pts) Develop a function named annualAdmissionsByCounty(year). This function will take a year as input, filter the data to identify all admissions for that specific year, and calculate the proportion of admissions by county. For instance, if the year is 2007, the function should filter the data for the specific year and compute the admissions as follows: county A 75%, county B 20%, and county C 2.5%, etc. The results should be displayed using a bar chart that highlights the top 10 counties.
Execute the annualAdmissionsByCounty() function in a new code cell to visualize the data for the last 10 years within the data.
Analyze the results and discuss any patterns or trends observed over this period.

```python
counties = list(set([county for county in df["County of Program Location"]]))

def annualAdmissionsByCounty(year):
    df_yr = df[df["Year"] == year]
    county_total_admissions = df_yr.groupby(["County of Program Location"])["Admissions"].sum()
    all_county_pct = (county_total_admissions / county_total_admissions.sum() * 100).sort_values(ascending=False)
    
    return all_county_pct
    
```

```python
years = [x + 2011 for x in range(1, 11)]
for year in years:
    all_county_pct = annualAdmissionsByCounty(year)

    ax = all_county_pct.head(10).plot(kind="bar", figsize=(9,4))
    ax.set_xlabel("Counties")
    ax.set_ylabel("Percentage of Admissions (%)")
    ax.set_title(f"{year} Share of Admissions for Top 10 counties")
    
    plt.show()
```

Unsurprisingly, the counties with the most admissions share was New York and Suffolk respectively, being 2 of the largest in the state. Nassau, Putnam, and Schenctady all alternate for the bottom in the top twn of a specific year, hovering around 3-4% of admissions. It's important to note that it's rarely one county dominating over the others, and most of times the top share is 13-14%.

```python
plt.bar(substances, substance_summations)

plt.xlabel("Substances")
plt.ylabel("Amount of Admitted Users (#)")
plt.title("Most used Substances at the Treatment Program")

plt.show()
```

Alcohol is clearly the most used substance, with Heroin being a close second. Most of the substances are relatively equal, implying that despite the wide range of people admitting to this program, there are similarites between most people.

```python
# per age group
admissions_by_age_substance = (filtered_df.groupby(['Age Group','Primary Substance Group'])['Admissions'].sum().unstack(fill_value=0).astype(int))
age_hash = admissions_by_age_substance.to_dict(orient='index')
print(age_hash)
```

```python
# Under 18
ax = admissions_by_age_substance.plot(kind='bar', stacked=True)

ax.set_xlabel('Age Group') 
ax.set_ylabel('Admissions')
ax.set_title('Rehab Admissions by Age Group and Primary Substance')
```

This stacked bar chart shows the admissions per age group and the substance group they were admitting for, giving us a comprehensive breakdown of what substance is most harmful. A significant portion og all substance abuse seems to be made up of alcohol and heroin, especially those aged 25-54. Other opids and marijuana seemed smaller in comparison, aside from those under 18.
