```python
df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')

print("\nInitial Data Info")
print(df.info())

print("\nFirst 10 rows of dataset")
print(df.head(10))

print("\nShape of dataset")
print(df.shape)

# Check for missing values
missing_values = df.isnull().sum().sum()
print(f"Total missing values: {missing_values}")

if missing_values > 0:
    df.dropna(inplace=True)
    print("Dropped rows with missing values.")

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\Duplicate Rows Detected: {duplicates}")

if duplicates > 0:
    df.drop_duplicates(inplace=True)
    print(f"Duplicates dropped. New Shape: {df.shape}")

# Convert Variables to Suitable Data Types
nominal_cols = [
    'Diabetes_012', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke',
    'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
    'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk', 'Sex'
]
ordinal_cols = ['GenHlth', 'Age', 'Education', 'Income']
discrete_cols = ['MentHlth', 'PhysHlth', 'BMI']

# Convert to Int64
df = df.astype('int64')

print("\nData Types Converted to Integers")
print(df.dtypes.head())

# Handle Outliers
print("\nOutlier Analysis")
invalid_days = df[(df['MentHlth'] > 30) | (df['MentHlth'] < 0) | (df['PhysHlth'] > 30) | (df['PhysHlth'] < 0)]
print(f"Rows with invalid Mental/Physical Health days (outside 0-30): {len(invalid_days)}")

# BMI Outlier Handling
Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"BMI Statistics: Mean={df['BMI'].mean():.2f}, Median={df['BMI'].median()}")
print(f"BMI Outlier Thresholds: < {lower_bound} or > {upper_bound}")

outliers = df[(df['BMI'] < lower_bound) | (df['BMI'] > upper_bound)]
print(f"Number of statistical outliers in BMI: {len(outliers)}")

# Remove extreme outliers
df= df[(df['BMI'] >= 12) & (df['BMI'] <= 80)]

# Final Data Summary
print("\nFinal Data Summary")
print(f"Final Shape: {df.shape}")
print(df.describe().T[['mean', 'std', 'min', 'max']])
```

### Question 3: EDA

```python
df['Diabetes_Label'] = df['Diabetes_012'].map({0: 'No Diabetes', 1: 'Pre-Diabetes', 2: 'Diabetes'})

def analyze_target_distribution(df):
    """Visualizes the imbalance in the dataset."""
    print("\n" + "="*50)
    print("ANALYSIS 1: TARGET VARIABLE DISTRIBUTION")
    print("="*50)

    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x='Diabetes_Label', data=df, order=['No Diabetes', 'Pre-Diabetes', 'Diabetes'], palette='viridis')

    # Add percentages on bars
    total = len(df)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom')

    plt.title('Distribution of Diabetes Classes', fontsize=15)
    plt.xlabel('Health Status')
    plt.ylabel('Count')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

    print("""
    INTERPRETATION:
    The dataset exhibits a significant class imbalance. The vast majority of observations (approx. 84%)
    represent individuals with 'No Diabetes'. The 'Pre-Diabetes' class is the smallest minority (~2%),
    while the 'Diabetes' class makes up the remainder.

    IMPLICATION FOR MODELING:
    This imbalance poses a challenge for machine learning. A model could achieve 84% accuracy simply by
    predicting everyone is healthy. We will likely need to employ resampling techniques (SMOTE, Undersampling)
    or use class-weighted metrics (F1-score, Precision-Recall) rather than simple accuracy to evaluate performance.
    """)

def analyze_correlations(df):
    """Heatmap to show relationship strength."""
    print("\n" + "="*50)
    print("ANALYSIS 2: CORRELATION MATRIX")
    print("="*50)

    # Drop the label column for correlation calculation
    corr_matrix = df.drop('Diabetes_Label', axis=1).corr()

    # Focus on correlation with the Target Variable
    target_corr = corr_matrix[['Diabetes_012']].sort_values(by='Diabetes_012', ascending=False)

    plt.figure(figsize=(14, 12))
    sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, fmt=".2f", linewidths=0.5)
    plt.title('Feature Correlation Matrix', fontsize=15)
    plt.tight_layout()
    plt.show()

    # Plot top correlations with Target
    plt.figure(figsize=(8, 8))
    sns.heatmap(target_corr, cmap='RdBu_r', annot=True, vmin=-1, vmax=1)
    plt.title('Correlation with Diabetes_012 Target', fontsize=15)
    plt.tight_layout()
    plt.show()

    print("""
    INTERPRETATION:
    Looking at the correlation heatmap, we can identify the strongest drivers of diabetes in this dataset.

    1. GenHlth (General Health): This shows the strongest positive correlation. Since GenHlth is an ordinal scale
       (1=Excellent to 5=Poor), this indicates that people reporting poorer general health are significantly more
       likely to be diabetic.
    2. HighBP (High Blood Pressure) and BMI: These are the next strongest physical indicators. The link between
       hypertension, obesity, and diabetes is well-documented medically and is clearly represented here.
    3. Age and HighChol: Positive correlations here confirm that diabetes risk increases with age and is often
       comorbid with high cholesterol.
    4. Inverse Correlations: Income and Education show negative correlations. This suggests that higher socioeconomic
       status is associated with a LOWER risk of diabetes, highlighting the impact of Social Determinants of Health.
    """)

def analyze_bmi_impact(df):
    """Violin plot to show density of BMI across classes."""
    print("\n" + "="*50)
    print("ANALYSIS 3: BMI DISTRIBUTION BY DIABETES STATUS")
    print("="*50)

    plt.figure(figsize=(12, 6))
    sns.violinplot(x='Diabetes_Label', y='BMI', data=df, order=['No Diabetes', 'Pre-Diabetes', 'Diabetes'], palette='muted')
    plt.title('BMI Distribution vs Diabetes Status', fontsize=15)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    print("""
    INTERPRETATION:
    The Violin Plot reveals distinct shape differences in BMI distribution:

    - No Diabetes: The distribution is heavily skewed towards the lower end (lower BMI), with the "bulb"
      of the violin situated around BMI 25-28.
    - Diabetes: The distribution shifts significantly upward. The median BMI is visibly higher, and the
      shape is fatter at higher BMI ranges (30-40+).
    - Pre-Diabetes: This class acts as a bridge, showing a distribution that is higher than the healthy group
      but slightly lower or comparable to the diabetic group.

    This confirms BMI as a critical separating feature. However, the overlap suggests BMI alone is not enough
    for classification; there are many overweight people without diabetes and some normal-weight people with it.
    """)

def analyze_risk_factors(df):
    """Normalized stacked bar charts for categorical risks."""
    print("\n" + "="*50)
    print("ANALYSIS 4: KEY RISK FACTORS (BP, CHOLESTEROL)")
    print("="*50)

    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Helper to calculate percentages
    def plot_normalized_stack(col, ax, title):
        ct = pd.crosstab(df[col], df['Diabetes_Label'], normalize='index')
        ct = ct[['No Diabetes', 'Pre-Diabetes', 'Diabetes']] # Reorder columns
        ct.plot(kind='bar', stacked=True, ax=ax, colormap='viridis')
        ax.set_title(title)
        ax.set_ylabel('Proportion')
        ax.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left')

    plot_normalized_stack('HighBP', axes[0], 'Diabetes Prevalence by High Blood Pressure (0=No, 1=Yes)')
    plot_normalized_stack('HighChol', axes[1], 'Diabetes Prevalence by High Cholesterol (0=No, 1=Yes)')

    plt.tight_layout()
    plt.show()

    print("""
    INTERPRETATION:
    These stacked bar charts normalize the data to show proportions, which is essential given the class imbalance.

    - HighBP: In the group with High Blood Pressure (1), the proportion of Diabetics increases drastically compared
      to the group without it. Approximately 60-70% of people with High BP are healthy, whereas nearly 95% of people
      WITHOUT High BP are healthy. Having High BP is a massive risk flag.
    - HighChol: A similar trend is visible. The presence of High Cholesterol roughly doubles or triples the likelihood
      of falling into the Diabetes or Pre-diabetes bucket compared to those with normal cholesterol.
    """)

def analyze_age_relation(df):
    """Age stratification analysis."""
    print("\n" + "="*50)
    print("ANALYSIS 5: AGE STRATIFICATION")
    print("="*50)

    # Age is ordinal 1-13. 1 is 18-24, 13 is 80+
    plt.figure(figsize=(14, 6))
    sns.kdeplot(data=df[df['Diabetes_Label'] == 'No Diabetes']['Age'], label='No Diabetes', shade=True)
    sns.kdeplot(data=df[df['Diabetes_Label'] == 'Pre-Diabetes']['Age'], label='Pre-Diabetes', shade=True)
    sns.kdeplot(data=df[df['Diabetes_Label'] == 'Diabetes']['Age'], label='Diabetes', shade=True)

    plt.title('Age Distribution by Diabetes Class (Density Plot)', fontsize=15)
    plt.xlabel('Age Category (1 = 18-24 ... 13 = 80+)')
    plt.legend()
    plt.show()

    print("""
    INTERPRETATION:
    The Density Plot (KDE) illustrates the probability of being in a specific age group for each health outcome.

    - The 'No Diabetes' curve (Blue) peaks earlier, suggesting younger people are far more likely to be healthy.
    - The 'Diabetes' curve (Green) is skewed heavily to the right. The prevalence of diabetes rises sharply
      starting around Age Category 8-9 (Age 55+).
    - This confirms Age is a non-negotiable factor. A model attempting to predict diabetes in a 20-year-old
      should behave differently than one predicting for an 80-year-old.
    """)

def analyze_socioeconomic(df):
    """Impact of Income and Education."""
    print("\n" + "="*50)
    print("ANALYSIS 6: SOCIAL DETERMINANTS OF HEALTH")
    print("="*50)

    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    sns.lineplot(x='Income', y='Diabetes_012', data=df, ax=axes[0], marker='o', color='teal')
    axes[0].set_title('Mean Diabetes Score vs Income Level')
    axes[0].set_ylabel('Mean Diabetes Risk Score (0-2)')

    sns.lineplot(x='Education', y='Diabetes_012', data=df, ax=axes[1], marker='o', color='purple')
    axes[1].set_title('Mean Diabetes Score vs Education Level')
    axes[1].set_ylabel('Mean Diabetes Risk Score (0-2)')

    plt.tight_layout()
    plt.show()

    print("""
    INTERPRETATION:
    Here we plot the 'Mean Diabetes Score' against Income and Education. Since the target is 0, 1, 2,
    a higher mean score implies a population with a higher prevalence of pre-diabetes and diabetes.

    - Both plots show a distinct downward trend. As Income increases (1 to 8), the risk of diabetes plummets.
    - As Education level rises (1 to 6), the risk also decreases.

    This validates the "Social Determinants of Health" theory mentioned in the prompt. Access to better food,
    healthcare, and health literacy (associated with higher income/education) are strong protective factors against diabetes.
    """)

# 2. Run Analyses
analyze_target_distribution(df)
analyze_correlations(df)
analyze_bmi_impact(df)
analyze_risk_factors(df)
analyze_age_relation(df)
analyze_socioeconomic(df)
```

```python
from scipy import stats
from scipy.stats import f_oneway
from scipy.stats import chi2_contingency
```

```python
df_test = df.sample(n=10000, random_state=42)
```

```python
# ANOVA: BMI across Diabetes_012 groups (0 = no, 1 = pre, 2 = diabetes)
groups = [df_test[df_test['Diabetes_012'] == g]['BMI'] 
          for g in sorted(df_test['Diabetes_012'].unique())]

f_stat, p_value = f_oneway(*groups)
print("ANOVA for BMI ~ Diabetes_012")
print(f"F-statistic: {f_stat:.3f}, p-value: {p_value:.3f}")

group_means = df_test.groupby('Diabetes_012')['BMI'].mean()
print("\nMean BMI by Diabetes_012:")
print(group_means)
```

The ANOVA test uses a continous and categorical variable to tell you if some difference exists between the groups. In this case, The extremely small p-val gives a definite difference between the groups of means of BMI and Diabetes. In Q3, the EDA plots already showed BMI going up as you move from No Diabetes → Pre → Diabetes. The mean BMI increased from X in non-diabetic to Y in diabetic patients and supports the EDA.

```python
def cramers_v(table):
    chi2, _, _, _ = chi2_contingency(table)
    n = table.to_numpy().sum()
    r, k = table.shape
    return np.sqrt(chi2 / (n * (min(r - 1, k - 1))))

# Chi-square: Age vs Diabetes_012
age_diab_table = pd.crosstab(df_test['Age'], df_test['Diabetes_012'])
chi2, p_value, dof, expected = chi2_contingency(age_diab_table)
cv = cramers_v(age_diab_table)

print("Chi-square for Age vs Diabetes_012")
print(f"Chi2: {chi2:.3f}, df: {dof}, p-value: {p_value:.3e}")
print(f"Cramer's V (effect size): {cv:.3f}")
```

In the chi-square, both variables are cotegorical and our test tests whether age and diabetes have a statistically significant effect on each other. We ran two tests, one for the aforementioned age and diabetes and the other for education and diabetes. To answer, the first, the p-value of below 0.05 means we reject the null hypothesis and older age groups are more likely to have diabetes.
